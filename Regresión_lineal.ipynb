```python
import seaborn as sns
import pandas as pd # Manejar datos
from sklearn.linear_model import LinearRegression # Hacer nuestro modelo
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score # Evaluar el modelo
import numpy as np # Manejar matrices
import matplotlib.pyplot as plt # Generar gráficas mucho más detalladas y avanzadas
```

```python
# Cargar datos. df = Dataframe
# .head para mostrar el encabezado con las dos primeras lineas

df = pd.read_excel('datos_producción.xlsx')
df.head(2)
```

```python
# Seleccionar variables, poner el mismo nombre que los nombres de las columnas

variable_x = "Horas Trabajadas"
variable_y = "Productos Terminados"
```

```python
# Generar análisis

modelo = LinearRegression()
modelo.fit(df[[variable_x]], df[[variable_y]]) # Las dos primmera slineas son utilizar una librería para que genere nuestro modelo de regresión
print ('Ecuación de la recta: y', round(modelo.coef_[0][0], 3), 'x +', round(modelo.intercept_[0], 3)) #Para obtener la ecuación de la recta. Access the first element of the arrays
print ('Coeficiente de correlación : ', round(np.corrcoef(df[variable_x], df[variable_y])[0,1], 3)) # Obtener el coeficiente de correlación
print ('Coeficiente de determinación:', round(r2_score(df[variable_y], modelo.predict(df[[variable_x]])), 3)) # Obtener el coeficiente de determinación

# Código para generar una gráfica de regresión con las variables y el intervalo de confianza en este caso al 95%

sns.regplot(x=df[variable_x], y=df[variable_y], data=df, ci=95, line_kws={'color': 'red'}, scatter_kws={'color': 'blue'})
plt.title('Regresión lineal simple con intervalo de confianza al 95%')
plt.xlabel(variable_x)
plt.ylabel(variable_y)
plt.show()

# El coeficiente de correación cuanto más se acerca a 1 o a -1 más fuerte es la correlación. En este caso, es
# positiva, por lo que cuando uno sube el otro también.
```

```python
# Generar predicción

dato_predictor = 15
# La cantidad que quieres predecir. En este caso pones las horas
# y predice cuanto sera la producción con esas horas trabajadas

horas_trabajadas_nuevas = pd.DataFrame({dato_predictor}, columns=[variable_x]) # Convertir a DataFrame
prediccion = modelo.predict(horas_trabajadas_nuevas) # Predecir
print(prediccion) # Imprimir la predicción
```
